```python
import pandas as pd

new_dict = {
    "pl_id": "",            
    "text1": "",
    "URL1": "",
    "text2": "",
    "URL2": "",
    "text3": "",
    "URL3": "",
}

df_input = pd.read_csv('./filtered_reivew.csv')

output_reivew = pd.DataFrame(new_dict,index=[0])
output_reivew_reivew_none = pd.DataFrame(new_dict,index=[0])
#############################################################################################
pl_ids = df_input['pl_id'].unique()

models = []
for pl_id in pl_ids:
    is_name = df_input['pl_id'] == pl_id
    model = df_input[is_name]
    
    
    if len(model)== 0:
        new_dict = {
            "pl_id": pl_id,            
            "text1": "",
            "URL1": "",
            "text2": "",
            "URL2": "",
            "text3": "",
            "URL3": "",
        }
        
    elif  len(model)== 1:
         new_dict = {
            "pl_id": pl_id,            
            "text1": model.iloc[0].text,
            "URL1": model.iloc[0].URL,
            "text2": "",
            "URL2": "",
            "text3": "",
            "URL3": "",
        }
        
    elif  len(model)== 2:
        new_dict = {
            "pl_id": pl_id,            
            "text1": model.iloc[0].text,
            "URL1": model.iloc[0].URL,
            "text2": model.iloc[1].text,
            "URL2": model.iloc[1].URL,
            "text3": "",
            "URL3": "",
        }
        
    elif  len(model)>= 3:
        new_dict = {
            "pl_id": pl_id,            
            "text1": model.iloc[0].text,
            "URL1": model.iloc[0].URL,
            "text2": model.iloc[1].text,
            "URL2": model.iloc[1].URL,
            "text3": model.iloc[2].text,
            "URL3": model.iloc[2].URL,
        }
    
    output_reivew = output_reivew.append(pd.DataFrame(new_dict,index=[0]))
```
